```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#reading the dataset
df = pd.read_csv("/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_84031.csv")
df
```

```python
#bar chart plotting
gender_counts = df['Region'].value_counts()
bar_width = 0.9
x=range(len(gender_counts.index))

plt.bar(gender_counts.index,gender_counts.values, facecolor='black')



plt.xlabel('Region')
plt.xticks(rotation=12, color='red')
plt.yticks(color='tab:green')
plt.ylabel('Count')
plt.title('Distribution of Region')







plt.xticks(x,gender_counts.index,rotation=45)
plt.tight_layout()
plt.show()
```
